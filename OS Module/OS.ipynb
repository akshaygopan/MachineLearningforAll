function to create a dictionary for header and value

```python
def create_data_dict(values,headers):
  result = {}
  for value,header in zip(values,headers):
    result[header] = value
  return result
```

read csv final function to read any csv file and make it into clean format

```python
def readcsv(path):
  result = []
  with open(path,'r') as f:
    lines = f.readlines()
    headers = parse_header(lines[0])
    for data_line in lines[1:]:
      values = parse_values(data_line)
      item_dict = create_data_dict(values,headers)
      result.append(item_dict)
    return result
```
